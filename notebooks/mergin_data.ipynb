```python
import pandas as pd
from pathlib import Path
import os

base_directory = Path(os.getcwd()).parent

weather_file = "data\\weather_data.csv"
weather_path = os.path.join(base_directory, weather_file)

energy_file = "data\\PL_data.csv"
energy_path= os.path.join(base_directory, energy_file)
```

```python
weather_df = pd.read_csv(weather_path, delimiter=",", parse_dates=["time"], index_col="time")
weather_df
```

```python
energy_df = pd.read_csv(energy_path, parse_dates=["utc_timestamp"], index_col="utc_timestamp")
energy_df.index.name = "time"
energy_df.index = energy_df.index.tz_convert(None)
energy_df
```

```python
combined_df = energy_df.join(weather_df, how="inner")
combined_df
```

```python
combined_df.columns
```

```python
name = "data\\data_for_training.csv"
full = os.path.join(base_directory, name)
combined_df.to_csv(full)
```
