```python
import pandas as pd
from pathlib import Path
import os

base_directory = Path(os.getcwd()).parent

weather_file = "data\\weather_cracow.csv"
weather_path = os.path.join(base_directory, weather_file)

energy_file = "data\\PL_data.csv"
energy_path= os.path.join(base_directory, energy_file)
```

```python
weather_df = pd.read_csv(weather_path, delimiter=",", skiprows=3, parse_dates=["time"], index_col="time")
weather_df
```

```python
energy_df = pd.read_csv(energy_path, parse_dates=["utc_timestamp"], index_col="utc_timestamp")
energy_df.index.name = "time"
energy_df.index = energy_df.index.tz_convert(None)
energy_df
```
